# Combine Datasets
Load the three cleaned CSV files and combine them into a single dataset.

```python
import pandas as pd
```

```python
# input files
files = [
    "ultrasonic_not_present_clean.csv",
    "ultrasonic_sedentary_clean.csv",
    "ultrasonic_active_clean.csv"
]
```

```python
dfs = []

for file in files:
    print(f"Loading {file}")
    df = pd.read_csv(file)
    dfs.append(df)
```

```python
# Combine all datasets
combined_df = pd.concat(dfs, ignore_index=True)

# Save combined dataset
combined_df.to_csv("ultrasonic_combined_clean.csv", index=False)

print("Combined dataset saved as ultrasonic_combined_clean.csv")
print("Total rows:", len(combined_df))
print("Label distribution:")
print(combined_df["label"].value_counts())
```
