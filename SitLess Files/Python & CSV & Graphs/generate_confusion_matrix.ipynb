# Confusion Matrix Visualisation
Generate a confusion matrix heatmap from the model evaluation results.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# confusion matrix values from model evaluation
# rows = actual, cols = predicted  [Active, Sedentary]
cm = np.array([[21, 3],
               [4,  19]])
```

```python
fig, ax = plt.subplots(figsize=(5.5, 4.5))

cmap = plt.cm.Blues
im = ax.imshow(cm, interpolation='nearest', cmap=cmap, vmin=0, vmax=25)
plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

classes    = ['Active', 'Sedentary']
tick_marks = [0, 1]
ax.set_xticks(tick_marks);  ax.set_xticklabels(classes, fontsize=12)
ax.set_yticks(tick_marks);  ax.set_yticklabels(classes, fontsize=12)

# annotate each cell with its count
thresh = cm.max() / 2.0
for i in range(2):
    for j in range(2):
        colour = 'white' if cm[i, j] > thresh else 'black'
        ax.text(j, i, str(cm[i, j]),
                ha='center', va='center',
                fontsize=18, fontweight='bold', color=colour)

ax.set_xlabel('Predicted Label', fontsize=13, labelpad=8)
ax.set_ylabel('Actual Label',    fontsize=13, labelpad=8)
ax.set_title('Confusion Matrix', fontsize=14, fontweight='bold', pad=12)

plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=180, bbox_inches='tight', facecolor='white')
plt.show()
print("Saved confusion_matrix.png")
```
