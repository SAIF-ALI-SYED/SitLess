# Data Cleaning
Clean the three raw CSV files by removing startup instability, replacing out-of-range readings, and resetting timestamps.

```python
import pandas as pd
```

```python
# the three raw CSV files to clean
files = [
    "ultrasonic_not_present.csv",
    "ultrasonic_sedentary.csv",
    "ultrasonic_active.csv"
]
```

```python
CLEANED_FILES = []

for file in files:
    print(f"Cleaning {file}...")
    df = pd.read_csv(file)

    # 1) drop the first 2 seconds (sensor is unstable at startup)
    df = df[df["time_sec"] > 2.0]

    # 2) replace out-of-range readings with -1
    df.loc[df["distance_cm"] > 300, "distance_cm"] = -1

    # 3) reset timestamps so each file starts from 0
    df["time_sec"] = df["time_sec"] - df["time_sec"].iloc[0]

    cleaned_name = file.replace(".csv", "_clean.csv")
    df.to_csv(cleaned_name, index=False)

    CLEANED_FILES.append(cleaned_name)
    print(f"Saved {cleaned_name}")

print("\nCleaning complete.")
print("Cleaned files:", CLEANED_FILES)
```
