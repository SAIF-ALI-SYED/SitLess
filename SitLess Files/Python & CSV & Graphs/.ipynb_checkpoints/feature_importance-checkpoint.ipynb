# Feature Importance
Extract and display the feature importance scores from the trained Random Forest model.

```python
import pandas as pd
import joblib
```

```python
# load the trained model and feature names
MODEL_FILE  = "rf_activity_model.pkl"
FEATURE_CSV = "ultrasonic_features.csv"

model = joblib.load(MODEL_FILE)

df = pd.read_csv(FEATURE_CSV)
feature_names = [col for col in df.columns if col != "label"]
```

```python
# extract and display feature importances ranked highest to lowest
importances = model.feature_importances_
importance_df = pd.DataFrame({
    "feature":    feature_names,
    "importance": importances
}).sort_values("importance", ascending=False)

print("Feature Importances:")
print(importance_df.to_string(index=False))
```
