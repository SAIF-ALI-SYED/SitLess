# Data Logging
Collect labeled ultrasonic distance data from the ESP32 sensor via serial connection.

**Requires:** ESP32 connected via USB (COM4).

```python
!pip install pyserial
import serial
import time
import csv
```

```python
# --- config (change these per session) ---
PORT          = "COM4"           # change if needed
BAUD          = 115200
LABEL         = "active"         # change per session
DURATION      = 120              # how many seconds to record
OUTPUT_FILE   = f"ultrasonic_{LABEL}.csv" 
```

```python
ser = serial.Serial(PORT, BAUD, timeout=1)
time.sleep(2)  # allow ESP32 to reset

print("Waiting for DIST data from ESP32...")

# wait for the first valid distance line before starting
while True:
    line = ser.readline().decode(errors="ignore").strip()
    if line.startswith("DIST:"):
        print("First valid reading received.")
        break
```

```python
start_time = time.time()

with open(OUTPUT_FILE, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["time_sec", "distance_cm", "label"])

    while time.time() - start_time < DURATION:
        line = ser.readline().decode(errors="ignore").strip()
        print("RAW:", line)

        if line.startswith("DIST:"):
            try:
                distance  = int(line.split(":")[1])
                timestamp = round(time.time() - start_time, 2)

                writer.writerow([timestamp, distance, LABEL])
                print(timestamp, distance, LABEL)

            except ValueError:
                pass

ser.close()
print(f"Saved to {OUTPUT_FILE}")
```
