# Distance vs Time Plot
Visualize the ultrasonic distance readings over time for all activity classes.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load combined cleaned dataset
df = pd.read_csv("ultrasonic_combined_clean.csv")
```

```python
plt.figure(figsize=(10, 5))

for label in ["not_present", "sedentary", "active"]:
    subset = df[df["label"] == label]
    plt.plot(
        subset["time_sec"],
        subset["distance_cm"],
        label=label,
        alpha=0.7
    )

plt.xlabel("Time (seconds)")
plt.ylabel("Distance (cm)")
plt.title("Ultrasonic Distance vs Time (Cleaned Dataset)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
