# Distance Distribution Boxplot
Compare the distance distributions of sedentary vs active classes.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ultrasonic_combined_clean.csv")

# Remove -1 for boxplot clarity
df_plot = df[df["distance_cm"] >= 0]

data = [
    df_plot[df_plot["label"] == "sedentary"]["distance_cm"],
    df_plot[df_plot["label"] == "active"]["distance_cm"]
]
```

```python
plt.figure(figsize=(6, 5))
plt.boxplot(data, labels=["Sedentary", "Active"])
plt.ylabel("Distance (cm)")
plt.title("Distance Distribution by Activity Class")
plt.grid(True)
plt.tight_layout()
plt.show()
```
